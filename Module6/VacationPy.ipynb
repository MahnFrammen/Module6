```python
# Dependencies and Setup
import hvplot.pandas
import pandas as pd
import requests
from pprint import pprint

# Import API key
from api_keys import geoapify_key
```

```python
# Load the CSV file created in Part 1 into a Pandas DataFrame
city_data_df = pd.read_csv("output_data/cities.csv")
city_data_df = city_data_df.rename(columns={"lon": "Lon"})

# Display sample data
city_data_df.head()
```

```python
import cartopy.crs as ccrs
import geoviews as gv
import pyproj

# Configure the map plot
citymap = city_data_df.hvplot.points("Lon", "Lat", 
                                     geo = True, 
                                     tiles = "OSM",
                                     size = "Humidity",
                                     scale = 0.8,
                                     color = "City")
citymap
```

```python
# Narrow down cities that fit criteria and drop any results with null values
city_data_df = city_data_df[(city_data_df["Max Temp"] <= 27) & (city_data_df["Max Temp"] >= 20)]
city_data_df = city_data_df[city_data_df["Wind Speed"] <= 5]
city_data_df = city_data_df[city_data_df["Cloudiness"] == 0]

city_data_df = city_data_df.dropna()
print(len(city_data_df))
city_data_df
```

```python
# Use the Pandas copy function to create DataFrame called hotel_df to store the city, country, coordinates, and humidity
hotel_df = city_data_df[["City", "Country", "Lat", "Lon", "Humidity"]].copy()

# Add an empty column, "Hotel Name," to the DataFrame so you can store the hotel found using the Geoapify API
hotel_df["Hotel Name"] = ""

hotel_df
```

```python
params = {"filter": f"circle:39.9086,-16.2325,10000",
              "bias": f"proximity:39.9086,-16.2325",
              "categories": "accommodation.hotel",
              "apiKey": geoapify_key}

name_address = requests.get("https://api.geoapify.com/v2/places", params=params).json()
pprint(name_address)
```

```python
radius = 10000
base_url = "https://api.geoapify.com/v2/places"

print("Starting hotel search")

for index, row in hotel_df.iterrows():
    params = {"filter": f"circle:{row['Lon']},{row['Lat']},{radius}",
              "bias": f"proximity:{row['Lon']},{row['Lat']}",
              "categories": "accommodation.hotel",
              "apiKey": geoapify_key}
    
    name_address = requests.get(base_url, params=params).json()

    # Grab the first hotel from the results and store the name in the hotel_df DataFrame
    try:
        hotel_df.loc[index, "Hotel Name"] = name_address["features"][0]["properties"]["name"]
    except (KeyError, IndexError): 
        hotel_df.loc[index, "Hotel Name"] = "No hotel found"
        
    print(f"{hotel_df.loc[index, 'City']} - nearest hotel: {hotel_df.loc[index, 'Hotel Name']}")

hotel_df
```

```python
import pandas as pd
import hvplot.pandas

hotel_df_cleaned = hotel_df.copy()
hotel_df_cleaned['Humidity'].fillna(0, inplace=True)
hotel_df_cleaned['Humidity'] = hotel_df_cleaned['Humidity'].astype(int)
hotel_df_cleaned = hotel_df_cleaned.dropna(subset=['Lon', 'Lat'])

citymap = hotel_df_cleaned.hvplot.points("Lon", "Lat", 
                                        geo=True, 
                                        tiles="OSM",
                                        hover_cols=["Country", "Hotel Name"],
                                        size="Humidity",
                                        color="City",
                                        xlabel="Longitude",
                                        ylabel="Latitude",
                                        title="Potential Vacation Spots"
                                       )
citymap
```
